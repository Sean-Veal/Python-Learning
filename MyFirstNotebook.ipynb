```python
watched_anime = 0
total_anime = 10

while True:
    # String formatting
    print("Anime {} watched".format(watched_anime+1))
    watched_anime += 1
    if watched_anime == total_anime:
        print("You've watched all the anime")
        break
```

# 7. Classes

```python
# Creating a class
class Person():
    # Constructor
    def __init__(self, name, age, color):
        # Create attributes
        self.name = name
        self.age = age
        self.color = color

    # Date of birth method
    def year_of_birth(self, current_year = 2025):
        return current_year - self.age

    #Projected age
    def project_age(self, years=5):
        return self.age + years
```

```python
new_person = Person("Sean", 33, "Blue")
```

```python
new_person.name
```

```python
result = new_person.year_of_birth()
print(result)
```

```python
special = Person("Fish", 65, "Blue")
```

```python
special.year_of_birth(current_year = 2023)
```

```python
print(new_person.project_age())
```

# 7. Inheritance

```python
class Astronaut(Person):
    #init function
    def __init__(self, name, age, color, mission_lengths_in_months):
        super().__init__(name, age, color)
        self.mission_lengths_in_months = mission_lengths_in_months
```

```python
astronaut = Astronaut("Sean", 33, "Blue", 48)
```

```python
astronaut.name
```
